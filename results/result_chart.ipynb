```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_table(df, title, figsize=(10, 4), fontsize=12):
    fig, ax = plt.subplots(figsize=figsize)
    ax.axis('off')

    table = ax.table(
        cellText=df.values,
        colLabels=df.columns,
        loc='center',
        cellLoc='center'
    )

    table.auto_set_font_size(False)
    table.set_fontsize(fontsize)
    table.scale(1, 1.4)

    ax.set_title(title, fontsize=14, pad=12)
    plt.tight_layout()
    plt.show()
```

```python
binary_df = pd.DataFrame({
    "Model": [
        "ChemBERTa",
        "GROVER",
        "RESCAL",
        "R-GCN",
        "R-GCN + Chem",
        "R-GCN + GROVER"
    ],
    "AUROC": [0.9504, 0.9435, 0.8321, 0.8579, 0.8878, 0.8500],
    "AUPR":  [0.9525, 0.9455, 0.7869, 0.8746, 0.9024, 0.8643]
})

plot_table(binary_df, "Binary Polypharmacy Prediction")
```
